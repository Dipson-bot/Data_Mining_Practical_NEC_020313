```python
import pandas as pd
```

```python
from pathlib import Path
dataset_path = Path.cwd() / "bread basket.csv"
df = pd.read_csv(dataset_path)
df.head()
```

```python
df.fillna("!", inplace=True)
df.head()
```

```python
dataset_list = []
for row in df.to_numpy().tolist():
    dataset_list.append([item for item in row if item != "!"])
```

```python
from mlxtend.preprocessing import TransactionEncoder

encoder = TransactionEncoder()
# Convert all items to strings before fitting
trfm = encoder.fit_transform([[str(item) for item in transaction] for transaction in dataset_list])
df = pd.DataFrame(trfm, columns=encoder.columns_)
df.head()
```

```python
from mlxtend.frequent_patterns import apriori, association_rules

freq_sets_df = apriori(df, min_support=0.0012, use_colnames=True)
```

```python
association_rules(freq_sets_df, min_threshold=0.12)
```
