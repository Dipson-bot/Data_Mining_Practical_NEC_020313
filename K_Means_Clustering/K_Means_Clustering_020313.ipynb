```python
import numpy as np
import pandas as pd
```

```python
from pathlib import Path

dataset_path = Path.cwd() / "gapminder.csv"
df = pd.read_csv(dataset_path)
df.head()
```

```python
X = df.drop(["Country", "Region"], axis=1) # Pass the column names as a list to the drop method. Use axis=1 to specify that you are dropping columns
```

```python
# Select only numeric columns
X = X.select_dtypes(include=np.number)

from sklearn.cluster import KMeans

X_arr = X.to_numpy()

# Replace infinite values with the maximum finite value
X_arr = np.nan_to_num(X_arr, posinf=np.finfo(np.float64).max)

model = KMeans(n_clusters=4)
model.fit(X_arr)

clusters = model.labels_
```

```python
columns = list(X.columns) # Use columns from X, which has the dropped columns removed
columns.append("cluster")
clusters_df = pd.DataFrame(np.hstack((X_arr, clusters.reshape(-1, 1))), columns=columns)
clusters_df["Country"] = df["Country"]
clusters_df["Region"] = df["Region"]
clusters_df
```
